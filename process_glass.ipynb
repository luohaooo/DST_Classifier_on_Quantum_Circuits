```python
import pandas as pd
import numpy as np
from simulator import conduct_simulation

# Experiments

# Input data processing
df = pd.read_table('glass.data', header=None)
raw_data = np.array(df)
pro_data = [[] for i in range(6)]
for item in raw_data:
    pro_item = str(item[0]).split(',')
    pro_item = [eval(pro_item[i]) for i in range(3,5)]+[pro_item[-1]]
    if pro_item[2] == '1':    
        pro_item[2] = 0
        pro_data[0].append(pro_item)
    if pro_item[2] == '2':    
        pro_item[2] = 1
        pro_data[1].append(pro_item)
    if pro_item[2] == '3':    
        pro_item[2] = 2
        pro_data[2].append(pro_item)
    if pro_item[2] == '5':    
        pro_item[2] = 3
        pro_data[3].append(pro_item)
    if pro_item[2] == '6':    
        pro_item[2] = 4
        pro_data[4].append(pro_item)
    if pro_item[2] == '7':    
        pro_item[2] = 5
        pro_data[5].append(pro_item)

# Components for GMM model
components_num = 3

# Proportion of training and test sets
train_ratio_list = np.array([0.3,0.4,0.5,0.6,0.7,0.8,0.9])

# Monte Carlo experiments num
N = 1

# Simulation
mean_accuracy_classical, mean_accuracy_quantum_32, mean_accuracy_quantum_1024 = conduct_simulation(pro_data, train_ratio_list, components_num, N)
```
